This is a notebook for getting commuting time by google map. Using the data we've got in other notebook 'forG', we will scrape google map by Selenium


```python
import urllib
from bs4 import BeautifulSoup
import urllib.request
import time
from selenium import webdriver
import pandas as pd
```

First, get data from 'forG.csv'

```python
forG = pd.read_csv("C:/Users/kenji/Downloads/forG.csv", engine='python')
forG.iloc[30]
```

```python
driver = webdriver.Firefox(executable_path=r"C:\\Users\kenji\Downloads\ADSproject\geckodriver.exe")
```

Open the google map and put the first destination coordinate (40.758895 -73.985131), Times Square, retreived at pm 9:50 Monday

```python
driver.get("https://www.google.com/maps/dir//1569-1551+Broadway,+New+York,+NY+10036/@40.7586759,-73.9866085,17z/data=!4m15!1m7!3m6!1s0x0:0x0!2zNDDCsDQ1JzMyLjAiTiA3M8KwNTknMDYuNSJX!3b1!8m2!3d40.758895!4d-73.985131!4m6!1m0!1m3!2m2!1d-73.9851389!2d40.7588889!3e3")
```

```python
x = len(forG)

md=[]
cg = '\n'
for i in forG.forG:
    driver.find_element_by_css_selector("#sb_ifc50 > input:nth-child(1)").send_keys(i)
    driver.find_element_by_css_selector("#sb_ifc51 > input:nth-child(1)").send_keys("40.758895 -73.985131")
    driver.find_element_by_css_selector("#directions-searchbox-1 > button:nth-child(2)").click()
    time.sleep(3)
    elements = driver.find_elements_by_css_selector("#section-directions-trip-0 > div:nth-child(2) > div:nth-child(2) > div:nth-child(1) > div:nth-child(1)")
    content = "".join([element.text for element in elements])
    md.append(content)    
    time.sleep(1)
```

Next, same approach to get data for (40.70688 -74.011266), NYSE in financial district, retreived at pm 10:00 Monday

```python
x = len(forG)

mdf=[]
cg = '\n'
for i in forG.forG:
    driver.find_element_by_css_selector("#sb_ifc50 > input:nth-child(1)").send_keys(i)
    driver.find_element_by_css_selector("#sb_ifc51 > input:nth-child(1)").send_keys("40.70688 -74.011266")
    driver.find_element_by_css_selector("#directions-searchbox-1 > button:nth-child(2)").click()
    time.sleep(3)
    elements = driver.find_elements_by_css_selector("#section-directions-trip-0 > div:nth-child(2) > div:nth-child(2) > div:nth-child(1) > div:nth-child(1)")
    content = "".join([element.text for element in elements])
    mdf.append(content)    
    time.sleep(3)
```

```python
dfmd = pd.Series(md,name='TimesSquare')
dfmdf = pd.Series(mdf,name='NYSE')
forGver2 = pd.concat([dfmd, dfmdf, forG], axis=1)
#forGver2.rename({0:'TimesSquare',1:'NYSE'},inplace=True)
```

We got commuting time data. So, again I will pass this data to notebook on ADRF for my convenience.
